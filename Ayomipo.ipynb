```python
import matplotlib.pyplot as plt
from math import pi, sqrt
```

```python
Voltage=[-157.979 * i/50 for i in range(50, -1, -1)]
```

```python
Freq = [1/(2*pi * sqrt( 5e-9 * 1.3702e-3 * 3.6972e-9 * sqrt(1- V/0.6))) for V in Voltage]
```

```python
fig, myplot = plt.subplots()
myplot.plot(Voltage, Freq)

myplot.set(xlabel='voltage (mv)', ylabel='frequency (Hz)',
       title='Graph of Frequency against Voltage')
myplot.grid()
```
